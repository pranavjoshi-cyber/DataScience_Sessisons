```python
## Bar plot 
enrollment =  [104,112,256,178,126,180]
courses = ['C','Java','Python','R','PHP','Javascript']
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled Students')
plt.grid()
plt.bar(courses, enrollment, color = 'VIOLET')
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled Students')
plt.grid()
plt.barh(courses, enrollment, color = 'VIOLET')
```

```python
# plot chart
plt.figure(figsize = (10,10))
plt.pie(enrollment, labels = courses, autopct = '%2.2f', shadow = True)
```

```python
# univariate analysis
# histogram
marks = np.array([67,56,45,61,93,72,65,49,68,62])
plt.title('Frequency plot marks')
plt.xlabel ('Range of marks')
plt.ylabel('Count')
plt.grid()
plt.xticks(range(0,101,10))
plt.hist(marks, bins = range(0,101,10), color = 'orange')
```

```python
# box plot 
plt.boxplot(marks )
```

```python
## machine learning 
# regression 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C://Users//Pranav//DataScienceSessions//Salary_Data.csv')
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
```

```python
#seperate the input and output data
x = df['YearsExperience'].values
y = df['Salary']
```

```python
x
```

```python
y;
```

```python
x = x.reshape(30,1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# creates the object of linear regression    
regressor = LinearRegression()
```

```python
# train the algorithm
regressor.fit(x,y)
```

```python
# Predict on the unseen data 
regressor.predict([[15]])
```

```python
regressor.predict([[20]])
```

```python
regressor.score(x,y)
```
