```python
# multiple regression 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import the dataset
df = pd.read_csv('mtcars.csv')
```

```python
salary = pd.read_csv('Salary_Data.csv')
```

```python
# seperate the input and output variables
x = df[['disp','hp','wt']]
y = df['mpg']
```

```python
x
```

```python
y
```

```python
# import the class
from sklearn.linear_model import LinearRegression
```

```python
# create the object 
reg = LinearRegression()
```

```python
# train the model
reg.fit(x,y)
```

```python
# predict on unseen data
disp = 221
hp = 102
wt = 3.51
reg.predict([[disp, hp, wt]])
```

```python
# visualize 
plt.figure(figsize = (16,9))
plt.subplot (2,2,1)
plt.scatter(df['disp'], y,color = 'r')
plt.subplot(2,2,2)
plt.scatter(df['hp'], y, color = 'g')
plt.subplot(2,2,3)
plt.scatter(df['wt'], y, color = 'b')
```

```python
reg.score(x,y)
```

```python
# decision tree classificiation 
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df
```

```python
x = df[['Age', 'EstimatedSalary']]
y = df[['Purchased']]
```

```python
x
```

```python
# explore the data
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary']) #c=y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# create the object
classifier = DecisionTreeClassifier(random_state = 0)
```

```python
#train the model
classifier.fit(x,y)
```

```python
#plot the tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12,15))
plot_tree(classifier, fontsize = 7, feature_names = ['age','sal'], class_names = ['y','n'], filled = True, rounded = True);
```

```python
#prediction on unseen data
new = pd.DataFrame({
    'Age': [25,37,56,66],
    'EstimatedSalary':[34000,41000,26000,134000]
})
```

```python
new
```

```python
classifier.predict(new)
```

```python
# naive bayes classification
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
```

```python
# train 
classifier.fit(x,y)
```

```python
classifier.predict(new)
```
